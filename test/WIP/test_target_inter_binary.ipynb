```python
fig, axes = plt.subplots(1, 2, gridspec_kw={'width_ratios':[2, 1]}, figsize=(8, 1))

cmap = plt.get_cmap('Blues')
norm = mpl.colors.Normalize(vmin=0., vmax=1.)

# create an inset axe in the current axe:
cax = inset_axes(axes[0],
                      height="20%", # set height
                      width="100%", # and width
                      loc=10) # center, you can check the different codes in plt.legend?

#cax = plt.axes([.2, .4, .35, .2])
cb1 = mpl.colorbar.ColorbarBase(cax, cmap=cmap,
                                norm=norm,
                                orientation='horizontal')
axes[0].set_xlabel('Average target value')
cb1.outline.set_linewidth(0)

axes[1].scatter(range(3), [1] * 3, s=[100, 500, 1000], edgecolors=cmap(1.), color='white')
axes[1].set_xlim(-1, 3)
axes[1].set_xlabel('Counts')

for i in range(2):
    for d in ['top', 'bottom', 'right', 'left']:
        axes[i].spines[d].set_visible(False)
    axes[i].set_xticks([])
    axes[i].set_yticks([])
```
